```python
import pandas as pd
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
cid ='37de602a7103451bbee4968209af0d87'
secret ='bd2179cfafdb4acbb4b53937511e8194'

client_credentials_manager = SpotifyClientCredentials(client_id=cid, client_secret=secret)
sp = spotipy.Spotify(client_credentials_manager = client_credentials_manager)
```

```python
path = "/Users/yobahbertrandyonkou/Downloads/NJU_MusicMood_v1.0/Relaxed/Test"
angry = pd.read_table(f"info.txt", sep=":", names=["index", "title", "artist", "duration"])
angry.head()
```

```python
# retreiving song details from spodify
rows = []
count = 0
for index, artist, title in tuple(angry[['index', 'artist', 'title']].values):
    query = f"artist:{artist} track:{title}"\
    .replace("(If Youre Wondering If I Want You To)", "")\
    .replace("(Album Version)", "").replace("(I'Ll Never Be)", "")\
    .replace("(Asereje)", "").replace("(Lp Version)", "").replace("(In Waves)", "")\
    .replace("(Sweet Dreams)", "")
#     print(query)
    sp_items = sp.search(query, limit=1)['tracks']['items']
    
    if len(sp_items) != 0:
        count += 1
        print(count, end=' ')
        song_details = sp_items[0]
        features = [
            index,
            song_details['id'],
            song_details['uri'],
            song_details['name'],
            song_details['album']['artists'][0]['name'],
            song_details['duration_ms'],
            song_details['explicit'],
            song_details['track_number'],
            song_details['album']['release_date'],
            song_details['popularity']
        ]
        rows.append(features)
```

```python
angry_train_df = pd.DataFrame(rows, columns=['index', "id", "uri", "title", "artist", "duration_ms", "explicit", "track_number", "release_date", "popularity"])
angry_train_df.head()
```

```python
# getting additional information about the track
def get_track_info(df):
    count = 0
    danceability = []
    energy = []
    key = []
    loudness = []
    speechiness = []
    acousticness = []
    instrumentalness = []
    liveness = []
    valence = []
    tempo = []
    mode = []
    uri = []
    for i in df['uri']:
        count += 1
        print(count, end = ' ')
        x = sp.audio_features(tracks=[i])[0]
        danceability.append(x['danceability'])
        energy.append(x['energy'])
        key.append(x['key'])
        loudness.append(x['loudness'])
        speechiness.append(x['speechiness'])
        acousticness.append(x['acousticness'])
        instrumentalness.append(x['instrumentalness'])
        liveness.append(x['liveness'])
        valence.append(x['valence'])
        tempo.append(x['tempo'])
        mode.append(x['mode'])
        uri.append(i)
            
            
    df2 = pd.DataFrame({
    'uri': uri,
    'danceability':danceability,
    'energy':energy,
    'key':key,
    'loudness':loudness,
    'speechiness':speechiness,
    'acousticness':acousticness,
    'instrumentalness':instrumentalness,
    'liveness':liveness,
    'valence':valence,
    'mode': mode,
    'tempo':tempo})
    
    return df2
```

```python
track_info = get_track_info(angry_train_df)
track_info.head()
```

```python
angry_train = angry_train_df.merge(track_info, how="left", left_on="uri", right_on="uri")
angry_train
```

```python
# fetching lyrics
angry_train['raw_lyrics'] = [open(f"{path}/relaxed_{index}.txt").read() for index in angry_train['index']]
angry_train.head()
```

```python
angry_train.to_csv("relaxed_test.csv", index=False)
```
