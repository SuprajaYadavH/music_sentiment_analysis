```python
import pandas as pd
```

```python
dataset = pd.read_csv("../Datasets/use_this_cleaned_dataset_for_eda.csv")
dataset.drop("Unnamed: 0", axis=1, inplace=True)
dataset.head()
```

```python
stopwords = ('ooh','okay','lyric','just','can','let','yeah','hey','wanna', '내', 'I','ayy','hmm','mmm','woah','also', '(','...','수','합','할','널','해','네')
```

```python
def remove_stopwords(lyrics):
    for stopword in stopwords:
        lyrics = str(lyrics).replace(stopword, "")
```

```python
dataset['lyrics_no_sw'] = dataset['lemmatized_lyrics'].apply(remove_stopwords)
```
